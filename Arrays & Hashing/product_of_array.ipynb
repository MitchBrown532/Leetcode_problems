## Product of Array Except Self
#### Difficulty: Medium

Given an integer array `nums`, return an array answer such that `answer[i]` is equal to the product of all the elements of `nums` except `nums[i]`.

The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.

You must write an algorithm that runs in `O(n)` time and without using the division operation.

Constraints:
- `2 <= nums.length <= 105`
- `-30 <= nums[i] <= 30`
- The product of any prefix or suffix of nums is guaranteed to fit in a 32-bit integer.

Link: https://leetcode.com/problems/product-of-array-except-self/



```python
# Naive approach: Calculate product to the left & right of each number in the array.
# This passes runtests, but does not satisfy the solution when submitted.
def productExceptSelf(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        # Init list to size of nums array
        output = [1]*len(nums)

        for i, num in enumerate(nums):
            product = 1

            # Calculate product to the left
            for x in range(i):
                product *= nums[x]
            
            # Calculate product to the right
            for y in range(i+1, len(nums)):
                product *= nums[y]

            # Add result, continue loop.
            output[i] = product
        
        return output
```

```python
# Calculate product of entire array. Then divide by selected number to find product without self.
# Dividing by 0 creates problems, special cases are needed.

def productExceptSelf(self, nums):
        """
        :type nums: List[int]
        :rtype: List[int]
        """
        # Init list to size of nums array
        output = [1]*len(nums)
        product = 1

        # Calculate product of entire array - be cautious of zeros!
        for num in nums:
            if num != 0:
                product *= num

        # Divide selected non-zero integer and append to output
        zero_count = nums.count(0)

        # If no zeros - proceed normally
        if zero_count == 0:
            for i in range(len(nums)):
                output[i] = product // nums[i]

        # If 1 zero - must calculate only the output for that index (all others will be 0)
        elif zero_count == 1:
            zero_index = nums.index(0)
            for i in range(len(nums)):
                if i == zero_index:
                    output[i] = product
                else:
                    output[i] = 0

        # If >1 zero - all outputs will be 0
        else:
            output = [0] * len(nums)
            
        return output
```
