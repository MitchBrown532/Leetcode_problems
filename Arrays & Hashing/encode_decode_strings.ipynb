## Encode and Decode Strings
#### Difficulty: Medium

Design an algorithm to encode a list of strings to a single string. The encoded string is then decoded back to the original list of strings.


Constraints:
- 0 <= strs.length < 100
- 0 <= strs[i].length < 200
- strs[i] contains only UTF-8 characters.


Link to problem: https://leetcode.com/problems/encode-and-decode-strings/description/


```python
class Solution:

    def encode(self, strs: List[str]) -> str:
        # Init empty string
        encoded = ""

        # Use string length and a seperator - this will make decoding easy
        for s in strs:
                encoded += str(len(s)) + "*" + s # concat into one long string
        
        #print(encoded)
        return encoded
            
        

    def decode(self, s: str) -> List[str]:
        decoded = []
        i = 0

        # Loop until end of string
        while i < len(s):
                j = i

                # Adjust j as we parse until finding '*' (this handles cases where the length is more than 1 digit)    
                while s[j] != '*':
                        j += 1
        
                # Store the len of the selected string
                length = int(s[i:j])

                # Move i past '*' seperator (makes slicing easy)
                i = j+1

                # use i + len to find new j
                j = i + length

                # extract substring and append
                decoded.append(s[i:j])

                # adjust i (ensures no repeated strings & termination of the while loop)
                i = j

        # print(decoded)
        return decoded
```
