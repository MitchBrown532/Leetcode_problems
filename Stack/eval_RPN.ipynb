## Evaluate Reverse Polish Notation
#### Difficulty: Medium

You are given an array of strings `tokens` that represents a valid arithmetic expression in Reverse Polish Notation.

Return the integer that represents the evaluation of the expression.
- The operands may be integers or the results of other operations.
- The operators include `'+'`, `'-'`, `'*'`, and `'/'`.
- Assume that division between integers always truncates toward zero.


Constraints:
- 1 <= `tokens.length` <= 10<sup>4</sup>
- `tokens[i]` is either an operator: `"+"`, `"-"`, `"*"`, or `"/"`, or an integer in the range `[-200, 200]`.

Link to problem: https://leetcode.com/problems/evaluate-reverse-polish-notation/description/

```python
# Solution 1: Classic application of a stack. RPN allows for calculations from an array in linear time.
# Without a stack, this would require nested loops (O(n^2)). Much more efficient.
class Solution:
    def evalRPN(self, tokens: List[str]) -> int:
        # Init stack
        nums = []

        # Loop until array is empty
        for i in tokens:
            # For numbers
            if i not in ['+','-','/','*']:
                nums.append(int(i)) # easier than converting with each pop
                # print("Testing num append:", nums)

            # For operands
            else:
                # print("Testing eval reach:", nums)
                # Pop the operands - remember they come out in reverse order (hence b first then a)
                b = nums.pop()
                a = nums.pop()
                # print("Testing eval pop:", nums)

                # Choose exp, evaluate, then append new num
                if i == '+':
                    nums.append(a + b)
                elif i == '-':
                    nums.append(a - b)
                elif i == '*':
                    nums.append(a * b)
                else:
                    # Truncate toward 0 
                    # Note: Cant use // or floor(), they truncate toward negative infitity (-2.44 -> -3). Using int() is how to truncate toward 0 (-2.64 -> -2).
                    nums.append(int(a / b))  

                # print("Testing eval append:", nums)

        return nums[0]

```
