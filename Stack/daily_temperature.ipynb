## Daily Temperatures
#### Difficulty: Medium

Given an array of integers `temperatures` represents the daily temperatures, return an array `answer` such that `answer[i]` is the number of days you have to wait after the `i`<sup>th</sup> day to get a warmer temperature. If there is no future day for which this is possible, keep answer`[i] == 0` instead.

<strong>Example:</strong>
- Input: `temperatures = [73,74,75,71,69,72,76,73]`
- Output: `[1,1,4,2,1,1,0,0]`

Constraints:
- 1 <= temperatures.length <= 105
- 30 <= temperatures[i] <= 100

Link to problem: https://leetcode.com/problems/daily-temperatures/description/

```python
# Solution 1: Uses a monotonic stack (decreasing)
    # Essetially a sorted stack - anytime that order would be broken an operation should be made.
    # In this case, each time a warmer temp is found, it triggers pops & updates
    # This ensures each element is pushed/popped once (O(n) time and space complexity)

class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        # Init
        n = len(temperatures)
        i_stack = [] # Stack of indicies
        answer = [0]*n # Array for answers (same len as temps, init to 0)

        for i, t in enumerate(temperatures):
            # If warmer - keep looping until warmer temp is reached or i_stack is empty (warmer than all previous)
            while i_stack and t >temperatures[i_stack[-1]]:
                j = i_stack.pop()
                answer[j] = i-j
            
            i_stack.append(i)
        
        return answer
                
```
