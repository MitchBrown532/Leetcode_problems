## Generate Parentheses
#### Difficulty: Medium

You are given an integer `n`. Return all well-formed parentheses strings that you can generate with `n` pairs of parentheses.

<strong>Example 1</strong>:
- Input: `n = 1`
- Output: `["()"]`

<strong>Example 2</strong>:
- Input: `n = 3`
- Output: `["((()))","(()())","(())()","()(())","()()()"]`

Constraints:
- 1 <= `n` <= 8

Link to problem: https://leetcode.com/problems/generate-parentheses/description/

```python
# Solution 1: Unfortunately - hardcoding all 8 solutions was not an option. Runtime woulda been great tho
# Originally was focused on trying to use a array version of a stack to implement this solution - found recursion made far more sense.
# Completed with exemplary runtime.
class Solution:
    

    def generateParenthesis(self, n: int) -> List[str]:
        # Init value (potential solution), output (array of valid solutions), and open/close (# of opening/closing brackets)
        value = ""
        output = []
        open, close = 0, 0

        # Generate all strings possible strings - prune invalid solutions early
            # Must create only valid solutions, otherwise runtime goes crazy
                # Keep track of # of open and closed brackets - if closed > open, invalid and solution can be pruned.
        
            # How to break when all solutions have been found?
                # Rather than using a loop: Use recursion????????

        # Helper function
        def backtrack(value, open, close):
            # Base Case:
            if len(value) == 2*n:
                # Valid solution found - append and return
                output.append(value)
                return
            
            # Recursive Cases:
                # Case 1: add a '(' (if valid)
            if open < n:
                backtrack(value + '(', open + 1, close)

                # Case 2: add a ')' (if valid)
            if close < open:
                backtrack(value + ')', open, close + 1)

                # Case 3: No valid solutions - let the branch die (no code needed)

        backtrack('', open, close)
        return output
        
        
```
