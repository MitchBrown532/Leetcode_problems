## Valid Parentheses
#### Difficulty: Easy

You are given a string s consisting of the following characters: '(', ')', '{', '}', '[' and ']'.

The input string s is valid if and only if:
1. Every open bracket is closed by the same type of close bracket.
2. Open brackets are closed in the correct order.
3. Every close bracket has a corresponding open bracket of the same type.

Return true if s is a valid string, and false otherwise.
Constraints:
- 1 <= s.length <= 1000

Link to problem: https://leetcode.com/problems/valid-parentheses/

```python
# Solution 1: Very straightforward application of stacks.
# Completed with optimal run time.
class Solution:
    def isValid(self, s: str) -> bool:
        # Init stack
        values = []
        
        for char in s:
            # Append opening brackets (no checking required)
            if char in ["[","(","{"]:
                values.append(char)

            # If closing bracket found, values.pop() will return most recent addition.
            # If not the corresponding braket, immediately break & return false.
            elif len(values) != 0:
                if char == "]":
                    if values.pop() != "[":
                        return False
                elif char == ")":
                    if values.pop() != "(":
                        return False
                else:
                    if values.pop() != "{":
                        return False
                    
            # Case handling: Leading with a closing bracket
            else:
                return False
            
        # True only if values is empty
        return not values
```
