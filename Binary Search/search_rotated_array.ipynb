## Search in Rotated Sorted Array
#### Difficulty: Medium

There is an integer array nums sorted in ascending order (with distinct values).

Prior to being passed to your function, `nums` is possibly left rotated at an unknown index `k` (1 <= `k` < `nums.length`) such that the resulting array is `[nums[k], nums[k+1], ..., nums[n-1], nums[0], nums[1], ..., nums[k-1]]` (0-indexed). For example, `[0,1,2,4,5,6,7]` might be left rotated by 3 indices and become `[4,5,6,7,0,1,2]`.

Given the array `nums` after the possible rotation and an integer `target`, return the index of `target` if it is in `nums`, or `-1` if it is not in `nums`.

You must write an algorithm with `O(log n)` runtime complexity.

 

<strong>Example 1:</strong>

- Input: `nums` = `[4,5,6,7,0,1,2]`, `target` = 0
- Output: 4

<strong>Example 2:</strong>

- Input: `nums` = `[4,5,6,7,0,1,2]`, `target` = 3
- Output: -1

<strong>Example 3:</strong>

- Input: `nums` = `[1]`, `target` = 0
- Output: -1
 

Constraints:
- 1 <= `nums.length` <= 5000
- -10<sup>4</sup> <= `nums[i]` <= 10<sup>4</sup>
- All values of `nums` are unique.
- `nums` is an ascending array that is possibly rotated.
- -10<sup>4</sup> <= `target` <= 10<sup>4</sup>

Link to problem: https://leetcode.com/problems/search-in-rotated-sorted-array

```python
# Solution 1: Similar logic to previous Binary Search problems.
# Uses a pivot (the min) to choose which half of the sorted list to look at
# used find_min() from previous problem to make this much easier
# Performed with exceptional Runtime
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        # Edge case
        if len(nums) == 1:
            if target == nums[0]:
                return 0
            else:
                return -1

        # -------------- Find min (from previous problem) -----------------
        # Init l, r
        l, r = 0, len(nums)-1
        # Binary Seach
        while l < r:
            mid = (l+r)//2
            if nums[r] >= nums[mid]: # if greater on the right
                r = mid
            else:
                l = mid + 1        
        ind = l

        # --------------Choose Pivot (disregard irrelevant half) ---------------
        if nums[ind] <= target <= nums[-1]:
            l, r = ind, len(nums) - 1
        else:
            l, r = 0, ind - 1

        # -------------- Find target using sorted half of list -----------------

        # Binary Seach for index
        while l<r:
            mid = (l+r)//2
            if target > nums[mid]:
                l = mid + 1
            else:
                r = mid

        if nums[l] == target:
            return l
        
        return -1
```
