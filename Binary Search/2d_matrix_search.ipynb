## Search a 2D Matrix
#### Difficulty: Medium

You are given an `m x n` integer matrix matrix with the following two properties:
- Each row is sorted in non-decreasing order.
- The first integer of each row is greater than the last integer of the previous row.

Given an integer `target`, return `true` if `target` is in `matrix` or `false` otherwise.

You must write a solution in `O(log(m * n))` time complexity.

<strong>Example 1:</strong>



- Input: `matrix` = `[[1,3,5,7],[10,11,16,20],[23,30,34,60]]`, `target` = `3`
- Output: `true`

<strong>Example 2:</strong>



- Input: `matrix` = `[[1,3,5,7],[10,11,16,20],[23,30,34,60]]`, `target` = `13`
- Output: `false`
 

Constraints:
- `m` == `matrix.length`
- `n` == `matrix[i].length`
- 1 <= `m`, `n` <= 100
- -104 <= `matrix[i][j]`, `target` <= 104

Link to problem: https://leetcode.com/problems/binary-search/description/

```python
# Solution 1: Logic is straightforward - 2 Binary searches, first to determine row then column.
# Used binary search from previous solution and repurposed it for this problem.
# Completed with exemplary runtime and memory.
class Solution:
    def searchMatrix(self, matrix: List[List[int]], target: int) -> bool:
        
        row_mid, row_top, row_bot = 0, 0, len(matrix)-1
        col_mid, col_l, col_r = 0, 0, len(matrix[0])-1
        
        while row_top <= row_bot:
            row_mid = (row_top+row_bot)//2

            if target > matrix[row_mid][-1]: # Greater than largest row value
                row_top = row_mid + 1
            elif target < matrix[row_mid][0]: # Smaller than smallest row value
                row_bot = row_mid -1
            else: # Row Found
                break
            
        while col_l <= col_r:
            col_mid = (col_l+col_r)//2
            
            if target > matrix[row_mid][col_mid]:
                col_l = col_mid + 1
            elif target < matrix[row_mid][col_mid]:
                col_r = col_mid -1
            else:
                return True
        
        # If not found
        return False
```
