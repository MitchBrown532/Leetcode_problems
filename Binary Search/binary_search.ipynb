## Binary Search
#### Difficulty: Easy

Given an array of integers `nums` which is sorted in ascending order, and an integer `target`, write a function to search `target` in `nums`. If `target` exists, then return its index. Otherwise, return -`1`.

You must write an algorithm with `O(log n)` runtime complexity.

 

<strong>Example 1:</strong>
- Input: `nums = [-1,0,3,5,9,12]`, `target = 9`
- Output: 4
Explanation: 9 exists in nums and its index is 4

<strong>Example 2:</strong>
- Input: nums = `[-1,0,3,5,9,12]`, `target` = 2
- Output: -1
Explanation: 2 does not exist in `nums` so return -1
 

Constraints:
- 1 <= `nums.length` <= 104
- -104 < `nums[i]`, `target` < 104
- All the integers in `nums` are unique.
- `nums` is sorted in ascending order.

Link to problem: https://leetcode.com/problems/binary-search/description/

```python
# Solution 1: Simple implementation of Binary search - iterative example (as opposed to recursive)
# Requires sorted list, operates in O(n log(n))
# Slices list in half recursively until target is found (or not)
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        l, r = 0, len(nums) -1

        while l <= r:
            mid = (l+r)//2
            if nums[mid] == target:
                return mid
            elif nums[mid] < target:
                l = mid + 1
            else:
                r = mid-1
        
        return -1
```

```python
# Solution 2: Simple implementation of Binary search - Recursive this time
# Exact same logic and time+space complexity as above. Despite this, it performs FAR worse on leetcode. (in both runtime and memory)
# Recursion is almost always slower than an iterative alternative.
# Only use in cases where readability is more important than efficiency.
class Solution:
    def search(self, nums: List[int], target: int) -> int:
        def helper(l,r):
            # Base Case: Not found/Empty
            if l > r:
                return -1

            # Base case: Target found
            mid = (l+r)//2
            if nums[mid] == target:
                return mid
            
            # Recursive Step: 
            elif nums[mid] < target:
                return helper(mid+1, r)
            else:
                return helper(l,mid-1)

        return helper(0, len(nums)-1)
```
