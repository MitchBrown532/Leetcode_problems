## Container With Most Water
#### Difficulty: Medium

You are given an integer array `heights` where `heights[i] `represents the height of the `i`<sup>th</sup> bar.

You may choose any two bars to form a container. Return the maximum amount of water a container can store.

Constraints:
- 2 <= `height.length` <= 1000
- 0 <= `height[i]` <= 1000

Link to problem: https://leetcode.com/problems/container-with-most-water/description/

```python
# Solution 1: Rather intuitive - example of 2-pointer problem that MUST NOT use sorting.
# Utility of 2-pointers is emphasized here - naive approach would require nested loops (O(n^2)). Two pointer solution is linear time.

class Solution:
    def maxArea(self, heights: List[int]) -> int:
        # Init pointers (l,r) & output
        l, r, output = 0,len(heights)-1,0

        while l < r:
            # print("----------Testing-----------")
            # print("Original Output: ", output)
            # print("Original Heights: L:", heights[l], 'R:', heights[r])

            # Calculate Volume
            # height = min of the two bars, width = difference of l & r
            vol = (r-l) * min(heights[l], heights[r])

            # Update output
            if vol > output:
                output = vol
                
            # Increment/decrement the lower bar 
            # Volume is dictated by height * width. Width is smaller with each iteration - Height must become greater or else volume will shrink.
            # If equal - change both?? - Changing 1 or both pointers will work.
            # By changing only 1 pointer you can skip an extra else statement.
            # 1 less thing to evaluate = better runtime & memory (by 4ms & 0.05mb in this case)
            if heights[l] < heights[r]:
                l += 1
            else:
                r -= 1

            # elif heights[l] > heights[r]:
            #     r -= 1
            # else:
            #     l += 1
            #     r-= 1

            # print("Updates Output: ", output)
            # print("Updated Heights: L:", heights[l], 'R:', heights[r])
            # print("----------End Of testing------------")


        return output
        
```
